# Import libraries

```python
import snscrape.modules.twitter as sntwitter
import stylecloud
import pandas as pd
from IPython.display import Image
```

# Gather data

```python
query =  "(@gotabayar) lang:en until:2022-04-13 since:2022-03-01"
tweets = []
limit = 5000
```

```python
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
  if len(tweets)==limit:
    break
  else:
    tweets.append([ tweet.date, tweet.content]) 
```

```python
df = pd.DataFrame(tweets, columns=['Date', 'Tweet'])
```

```python
df['Tweet'].to_csv('data.csv', index = False)
```
